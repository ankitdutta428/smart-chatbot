```python
healthcare = pd.read_csv('/kaggle/working/healthcare.csv')
telecom = pd.read_csv('/kaggle/working/telecom.csv')
```

```python
healthcare = healthcare.drop('Description', axis=1)
healthcare.head(10)
```

```python
drop_columns = ['intent', 'category', 'tags']
telecom = telecom.drop(columns = drop_columns)
telecom.head(10)
```

```python
healthcare = healthcare.rename(columns={'Patient':'prompt', 'Doctor':'response'})
telecom = telecom.rename(columns={'instruction':'prompt', 'reponse':'response'})
banking = banking.rename(columns={'Query':'prompt', 'Response':'response'})
```

```python
healthcare.head(10)
```

```python
dataframes_with_info = [
    (healthcare, "HC", 0),
    (telecom, "TC", 1),
    (ecommerce, "EC", 2),
    (banking, "BA", 3),
]

# Initialize an empty list to store the processed dataframes
processed_dfs = []

# Iterate through each dataframe, adding the new columns
for df, category, category_id in dataframes_with_info:
    df['category'] = category
    df['category_id'] = category_id
    processed_dfs.append(df)


# Concatenate all processed dataframes into one
merged_df = pd.concat(processed_dfs, ignore_index=True)

merged_df.tail(10)
```

```python
category_counts = merged_df['category'].value_counts()

# Print the counts
print("Row counts per category:")
print(category_counts)
```

```python
rows_to_remove = {
    'HC': 256516, #replace with 254000 when testing on full dataset
    'TC': 25600 #replace with 22500 when testing on full dataset
}

# Function to delete rows randomly
def delete_random_rows(df, category, num_rows):
    df_category = df[df['category'] == category]
    if len(df_category) <= num_rows:
        print(f"Warning: There are only {len(df_category)} rows in category '{category}'. Deleting all rows")
        return df[df['category'] != category]
    else:
        indices_to_drop = np.random.choice(df_category.index, num_rows, replace=False)
        return df.drop(indices_to_drop)

# Delete rows for HC and TC
for category, num_rows in rows_to_remove.items():
    merged_df = delete_random_rows(merged_df, category, num_rows)

# Verify counts
category_counts_after_removal = merged_df['category'].value_counts()
print("\nRow counts per category after removal:")
print(category_counts_after_removal)
```

```python
merged_df.to_csv('final_training_data.csv')
```
